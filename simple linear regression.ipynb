```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df=pd.read_csv(r"D:\datasets\homeprices.csv")
df.head()
```

```python
plt.xlabel("Area(sq ft)")
plt.ylabel("Price($)")
plt.plot(df.Area,df.Price,color='red',marker='o')
```

```python
reg=LinearRegression()
reg.fit(df[['Area']],df.Price)
```

```python
reg.predict([[3300]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
#Let check from y=mx+c equation for linear regression condition and therefore we get same values
reg.coef_*3300 + reg.intercept_
```
